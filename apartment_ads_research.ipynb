# Исследование объявлений о продаже квартир

<b>Описание проекта</b>

В моем распоряжении данные сервиса Яндекс.Недвижимость — архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктов за несколько лет. Нужно научиться определять рыночную стоимость объектов недвижимости. Моя задача — установить параметры. Это позволит построить автоматизированную систему: она отследит аномалии и мошенническую деятельность. 

По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые — получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма. 

**Цель проекта**

Используя данные сервиса Яндекс.Недвижимость, определить рыночную стоимость объектов недвижимости и типичные параметры квартир.


<b>Описание данных</b>

- <b>airports_nearest</b> — расстояние до ближайшего аэропорта в метрах (м)
- <b>balcony</b> — число балконов
- <b>ceiling_height</b> — высота потолков (м)
- <b>cityCenters_nearest</b> — расстояние до центра города (м)
- <b>days_exposition</b> — сколько дней было размещено объявление (от публикации до снятия)
- <b>first_day_exposition</b> — дата публикации
- <b>floor</b> — этаж
- <b>floors_total</b> — всего этажей в доме
- <b>is_apartment</b> — апартаменты (булев тип)
- <b>kitchen_area</b> — площадь кухни в квадратных метрах (м²)
- <b>last_price</b> — цена на момент снятия с публикации
- <b>living_area</b> — жилая площадь в квадратных метрах (м²)
- <b>locality_name</b> — название населённого пункта
- <b>open_plan</b> — свободная планировка (булев тип)
- <b>parks_around3000</b> — число парков в радиусе 3 км
- <b>parks_nearest</b> — расстояние до ближайшего парка (м)
- <b>ponds_around3000</b> — число водоёмов в радиусе 3 км
- <b>ponds_nearest</b> — расстояние до ближайшего водоёма (м)
- <b>rooms</b> — число комнат
- <b>studio</b> — квартира-студия (булев тип)
- <b>total_area</b> — общая площадь квартиры в квадратных метрах (м²)
- <b>total_images</b> — число фотографий квартиры в объявлении

```python
# Напишем функцию, подсчитывающую координаты усов.
def moustache(df, column):
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    mx = df[column].max()
    mn = df[column].min()
    q = q3 - q1
    left_m = q1 - 1.5*q
    right_m = q3 + 1.5*q
    if mn > left_m:
        left_m = mn
    if mx <= right_m:
        right_m = mx
    return left_m, right_m
```

```python
# посчитали границы усов
col = 'total_area'
moustache(data, col)
```

```python
# строим гистограмму по границам усов
(
    data['total_area']
    .plot(kind='hist', bins= 60, grid=True, range=(19, 114.25), title='Расспределение квартир по общей площади')
)
plt.xlabel('Общая площадь квартиры, м²')
plt.ylabel('Количество квартир')
plt.show()
```

```python
#Посчитаем границы выбросов по общей стоимости квартиры
col = 'last_price'
moustache(data, col)
```

```python
#график зависимости стоимости объекта от того, на каком этаже находится объект по данным без выбросов
(
    data
    .query('last_price <= 11000000')
    .pivot_table(index = ['floor_type'], values=['last_price'])
    .plot(grid=True, figsize=(5, 5), style='o-', title='Зависимость стоимости объекта от типа этажа объекта')
)
plt.xlabel('Количество комнат объекта')
plt.ylabel('Стоимость объекта')
plt.show()
```

С менее дорогими квартирами дело обстоит так:
Самые дорогие квартиры в основном находятся между первым и последним этажами.
Самые дешевые на первом этаже.

Это достаточно логично:
- квартиры на первом этаже самые дешевые, лучше жить выше, но с лучшим видом.
- квартиры на последнем этаже не самые дорогие в старых домах, потому что есть вероятность того, что протечет крыша.
- самые дорогие квартиры на последнем этаже, если это пентхаусы например, но их доля в нашем датасете мала.


##### Дата размещения (день недели, месяц, год)

В иследовании зависимоти от даты размещения, сделаем среез по цене, чтобы убрать выбросы, аналочино прошлому пункту.
Так мы получим более правдивые данные.

```python
#график зависимости стоимости объекта от дня недели размещения объявления.
(
    data
    .query('last_price < 11000000')
    .pivot_table(index = ['weekday'], values=['last_price'])
    .plot(style='o-', grid=True, figsize=(5, 5), title='Зависимость стоимости объекта от дня недели размещения объявления')
)
plt.xlabel('День недели размещения объявления')
plt.ylabel('Стоимость объекта')
plt.show()
```

```python
#график зависимости стоимости объекта от месяца размещения объявления
(
    data
    .query('last_price < 11000000')
    .pivot_table(index = ['month'], values=['last_price'])
    .plot(style='o-', grid=True, figsize=(5, 5), title= 'Зависимость стоимости объекта от месяца размещения объявления')
)
plt.xlabel('Месяц размещения объявления')
plt.ylabel('Стоимость объекта')
plt.show()
```

Самые высокие цены - в апреле.
Самые низкие цены - в июне.

```python
#график зависимости стоимости объекта от года размещения объявления
(
    data
    .query('last_price < 11000000')
    .pivot_table(index = ['year'], values=['last_price'])
    .plot(style='o-', grid=True, figsize=(5, 5), title= 'Зависимость стоимости объекта от года размещения объявления')
)
plt.xlabel('Год размещения объявления')
plt.ylabel('Стоимость объекта')
plt.show()
```

#### **Выявление населённых пунктов с самой высокой и низкой стоимостью квадратного метра**

- Выделим населённые пункты с самой высокой и низкой стоимостью квадратного метра.

```python
top = data['locality_name'].value_counts().head(10)
top
```

```python
top_10 = (
    data
    .query('locality_name in @top.index')
    .pivot_table(index='locality_name', values='square_meter_price', aggfunc=['count', 'mean'])
)
top_10.columns = ['total_ad', 'square_meter_mean_price']
top_10['square_meter_mean_price'] = round(top_10['square_meter_mean_price']).astype(int)
top_10
```

```python
(
    top_10
    .plot(y='square_meter_mean_price', grid=True, figsize=(5, 5), kind= 'bar', title= 'Стоимость одного квадратного метра в разных населенных пунктах')
)
plt.xlabel('Название населенного пункта')
plt.ylabel('Средняя стоимость за м²')
plt.show()
```
